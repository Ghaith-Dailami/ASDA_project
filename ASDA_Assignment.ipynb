```python
#now focus on data and allxl
# merge the sheets into one
#start by creating an empty list, in which, you will add the sheet names
all_xl_sheets=[]
#now create a for loop in your data
for sheet_name in allxl.sheet_names:
    daf=allxl.parse(sheet_name)
    all_xl_sheets.append(daf)

#concatenate or merge all the sheets into one data frame
merged_file=pd.concat(all_xl_sheets, ignore_index=True)

#print the file
print ("Merged File:")
print (merged_file)
```

```python
#create a map for the cities and their countries, to add later to the new column called country
city_to_country={
    'barcelona':'Spain',
    'amsterdam':'Netherlands',
    'athens':'Greece',
    'berlin':'Germany',
    'paris':'France',
    'lisbon':'Portugal',
    'budapest':'Hungary',
    'london':'UK',
    'rome':'Italy',
    'vienna':'Austria'}
#now create an empty list so you can fill later
sheet_with_all_info=[]
#now use the for loop to loop over all the sheets and extract the city name and the weekends or weekdays info
for sheet_name in allxl.sheet_names:
    daaf=allxl.parse(sheet_name)

#create the new columns 
    daaf['Weekends/Weekdays']=''
    daaf['City']=''
    daaf['Country']=''

#separate the names of the sheets to extrace the info regarding the city and the weekend or weekdays
    parts=sheet_name.lower().split('_')
    city=parts[0]
    time=parts[1]

# now populate the columns

#starting with the info of Weekends/Weekdays
    daaf['Weekends/Weekdays']= 'Weekends' if 'weekends' in time else 'Weekdays'

# now populate the city column
    daaf['City']=city.capitalize()

# populate the country column using the info from city and using the mapping created above
    daaf['Country']=city_to_country.get(city, 'unknown')

#add the new data to our new dataset using the append function
    sheet_with_all_info.append(daaf)

#now merge the data 
final_merged_data_with_new_info=pd.concat(sheet_with_all_info, ignore_index=True)

#print the new data
print (final_merged_data_with_new_info)
```

```python
#Suggested questions for testing (for beginners)
#Is there a difference in price between two cities? Choose at least three pairs of cities during weekdays.
# start by filtering out the data that includes weekdays values only
weekday_data=fine_data[fine_data['Weekends/Weekdays']=='Weekdays']
#calculate the mean of city prices
weekday_prices=weekday_data.groupby("City")["Price"].mean()
#let us now group the cities together
city_groups=(['Rome','Barcelona'],['Berlin','Lisbon'],['Athens','Paris'])
for city_1, city_2 in city_groups:
    print (f'Average Price in {city_1}: {weekday_prices[city_1]}')
    print (f'Average Price in {city_2}: {weekday_prices[city_2]}')
    print (f"Difference in price between {city_1} and {city_2}:{abs(weekday_prices[city_1]-weekday_prices[city_2])}\n")
```

```python
#Does the price tend to be higher on weekends?
#get the price on weekends and weekdays (mean) and compare
price_on_weekends=fine_data[fine_data['Weekends/Weekdays']=='Weekends']['Price'].mean()
price_on_weekdays=fine_data[fine_data['Weekends/Weekdays']=='Weekdays']['Price'].mean()
print (f'The mean price on Weekends is {price_on_weekends}')
print(f'The mean price on Weekdays is {price_on_weekdays}')
if price_on_weekends>price_on_weekdays:
    print ('Yes, the price tends to be higher on Weekends')
else:
    print ('No, the price tends to be higher on Weekdays')
```

```python
#Are listings of superhosts more expensive than those of normal hosts?
#get prices of when the host is superhost and when they are normal hosts'
#make all the values strings
fine_data['host_is_superhost'] = fine_data['host_is_superhost'].astype(str)
#get th mean prices
superhost_price=fine_data[fine_data['host_is_superhost']=='True']['Price'].mean()
normalhost_price=fine_data[fine_data['host_is_superhost']=='False']['Price'].mean()

print (f'The mean price of superhosts is {superhost_price}')
print (f'The mean price of normalhosts is {normalhost_price}')

if superhost_price>normalhost_price:
    print ('Yes, superhost listings are more expensive than those of normal hosts')
else:
    print ('No, superhost listings are not more expensive than those of normal hosts')
```

```python
#Are superhosts closer to the city center and metro station than normal hosts?
#calculate the means of both types of hosts to city center
sh_dist_to_cc=fine_data[fine_data['host_is_superhost']=='True']['dist'].mean()
nh_dist_to_cc=fine_data[fine_data['host_is_superhost']=='False']['dist'].mean()

#now calculate the means of the distance to the metro
sh_dist_to_metro=fine_data[fine_data['host_is_superhost']=='True']['metro_dist'].mean()
nh_dist_to_metro=fine_data[fine_data['host_is_superhost']=='False']['metro_dist'].mean()


#print them out appropriately 
#starting with the distacne to the city center
print (f'The mean distance to City Center for superhosts is {sh_dist_to_cc}')
print (f'The mean distance to City Center for normal hosts is {nh_dist_to_cc}')
if sh_dist_to_cc<nh_dist_to_cc:
    print ('Yes, super hosts are closer to the city center compared to normal hosts')
else:
    print ('No, Super hosts are not closer to the city center compared to normal hosts')

#print the metro distances
print (f'The mean distance to the metro for superhosts is {sh_dist_to_metro}')
print (f'The mean distance to the metro for normal hosts is {nh_dist_to_metro}')
if sh_dist_to_metro<nh_dist_to_metro:
    print ('Yes, super hosts are closer to the metro compared to normal hosts')
else:
    print ('No, Super hosts are not closer to the metro compared to normal hosts')
```

```python
#Are superhosts cleaner than normal hosts?
#get the mean rating for superhosts and normal hosts and compare
sh_cleanliness=fine_data[fine_data['host_is_superhost']=='True']['cleanliness_rating'].mean()
nh_cleanliness=fine_data[fine_data['host_is_superhost']=='False']['cleanliness_rating'].mean()

print(f'The average cleanliness rating for superhosts is {sh_cleanliness}')
print(f'The average cleanliness rating for normalhosts is {nh_cleanliness}')

if sh_cleanliness>nh_cleanliness:
    print ('Yes, superhosts are cleaner than normal hosts')
else:
    print ('No, superhosts are not cleaner than normal hosts')
```

```python
import folium as fo
#list the cities available
cities=fine_data['City'].unique()
#loop on cities
for city in cities:
    city_data=fine_data[fine_data['City']==city]
    #sort for the most expensive locations
    top_10_expen=city_data.nlargest(10, 'Price')
    top_10_cheap=city_data.nsmallest(10, 'Price')
    #create a map and centre it on the average location of the listings
    city_map=fo.Map(location=[city_data['lat'].mean(),city_data['lng'].mean()],zoom_start = 12, tiles='cartodbpositron')
    #list the 10 most expensive
    for idx, row in top_10_expen.iterrows():
        fo.Marker(location=[row['lat'],row['lng']], popup=f"Price{row['Price']}-price {row['room_type']}", tooltip='Top 10 most expensive'
                  ,icon=fo.Icon (color='green',icon = 'glyphocon-usd')).add_to(city_map)
        
#create one for the cheap places
    for idx, row in top_10_cheap.iterrows():
        fo.Marker(location=[row['lat'],row['lng']], popup=f"Price{row['Price']}-price {row['room_type']}", tooltip='Top 10 cheapest'
                  ,icon=fo.Icon (color='blue',icon = 'glyphocon-usd')).add_to(city_map)

    # save each as an HTML file and then open them
    city_map.save(f'{city}_airbnb_map.html')
    print(f'Map for {city} saved as {city}_airbnb_map.html')
```
